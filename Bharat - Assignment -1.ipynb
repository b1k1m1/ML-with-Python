Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line.


```python
l= []

for i in range(2000,3201):
    if(i%7 == 0) and (i%5 != 0):
        l.append(str(i))
    
#print(l)
print (','.join(l))
```

Create the below pattern using nested for loop in Python.

 
* 
* * 
* * * 
* * * * 
* * * * * 
* * * * 
* * * 
* * 
* 

```python
for row in range(9):
    for col in range(5):
        if (row in {0,8}) and (col == 0):
            print('*', end = '')
        elif (row in {1,7}) and (col in {0,1}):
            print('*', end = '')
        elif (row in {2,6}) and (col in {0,1,2}):
            print('*', end = '')
        elif (row in {3,5}) and (col in {0,1,2,3}):
            print('*', end = '')
        elif row == 4:
            print ('*', end = '')
        else:
            print(" ", end = '')
    print()
```

Write a Python program to reverse a word after accepting the input from the user.

Sample Output: 

Input word: AcadGild

Output: dilGdacA

```python
input_string = input('Please enter a word for get reversed value :')
count = len(input_string)
result = ''
while count > 0:
    result = result + input_string[count -1]
    count = count - 1
print(result)
```
